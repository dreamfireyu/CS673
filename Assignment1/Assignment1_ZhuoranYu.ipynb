```python
# q2 multiplication table for numbers up to 12.
for i in range(1,13):
    for j in range(1,13):
        print(f"{i} * {j} = {i*j} ")
```

```python
# q3 determine palindrome
def check_palindrome(str):
    mid = int(len(str)/2)
    for i in range(mid):
        if str[i]!=str[-i-1]:
            return False
    return True

while True:
    str = input("Input a string:")
    if str=="q":
        break
    if check_palindrome(str)==True:
        print(f"The string <{str}> is a palindrome.")
    else:
        print(f"The string <{str}> is not a palindrome.")
```

```python
# q4 Compare two lists

line1 = input("Input the first list: ")
line2 = input("Input the second list: ")
list1 = line1[1:-1].split(',')
list2 = line2[1:-1].split(',')
if len(list1) == len(list2):
    res = []
    for i in range(len(list1)):
        res.append(list1[i])
        res.append(list2[i])
    print(f"Merging result is : {res}")
else:
    print("These two lists are not of equal length.")
```

```python
# q6 determine leap yaer.
while True:
    year = input("Input a year: ")
    if year=="q":
        break
    year = int(year)
    if (year%400==0) or ((year%4==0) and (year%100!=0)):
        print(f"Year {year} is leap year.")
    else:
        print(f"Year {year} is not leap year.")
```
